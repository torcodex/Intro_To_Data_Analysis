```python
emp_data = pd.read_csv("/content/drive/MyDrive/employee_data.csv")
emp_data
```

```python
# duplicated() used to find if there are duplicates
emp_data.duplicated()
```

```python
# These are the two methods for droping duplicates:
#    assin it so that teh action takes place correctly
emp_data = emp_data.drop_duplicates()
#    use the attribute "inplace" to avoid assigning
emp_data.drop_duplicates(inplace=True)
emp_data
```

```python
# isnull()/isna() is used to find the null values in each columns in data
# .sum() is stacked on top of it because we need to find sum of null values in each columns
emp_data.isnull()
emp_data.isna()
```

```python
emp_data.isnull().sum()
emp_data.isna().sum()
```

```python
# .loc used to tell pandas the exact location where the updation is taking place
emp_data.loc[(emp_data['Department']=="IT Tech")&(emp_data['Salary'].isna()),'Salary'] = emp_data['Salary'].mean()
emp_data
```

```python
# read_csv() can't read multiple sheets if it's on one csv file
# read_excel() can read multple sheets
```

```python
# used to remove rows with null value
# emp_data.dropna()
# Output:
#   Name	Age	Department	Salary	Join_Date
# 0	Rajesh	29.0	Ops	60000.0	21-05-2023
# 3	Gilbert	29.0	Ops	60000.0	21-05-2021
# 4	Kevin	28.0	Support	61000.0	30-07-2024
# 5	Navin	35.0	Support	72000.0	19-08-2023
```

```python
emp_data.loc[(emp_data['Age'].isna()),'Age'] = emp_data['Age'].mean()
emp_data
```

```python
emp_data.fillna("Unkown",inplace=True)
emp_data
```
