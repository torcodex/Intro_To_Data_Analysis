```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars_data = pd.read_csv('/content/drive/MyDrive/mtcars.csv')
cars_data
```

```python
cars_data.shape
```

```python
cars_data.info()
```

```python
cars_data.describe()
```

```python
# histogram is apt for showing continous data
plt.hist(cars_data['mpg'])
plt.xlabel('mpg')
plt.ylabel('Frequency')
plt.title('Histogram of mpg')
plt.show()
# it shows more people tend to buy cars with mpg in range (13-20) and it can be suggested to other users
```

```python
plt.hist(cars_data['disp'])
plt.xlabel('disp')
plt.ylabel('Frequency')
plt.title('Histogram of disp')
plt.show()
# here people tend to buy cars with disp of range(75 -140) or (275 - 325)
```

```python
plt.hist(cars_data['hp'])
plt.xlabel('HP')
plt.ylabel('Frequency')
plt.title('Histogram of HP')
plt.show()
# here people tend to buy cars with hp of range(120 -140)
```

```python
plt.hist(cars_data['drat'])
plt.xlabel('drat')
plt.ylabel('Frequency')
plt.title('Histogram of drat')
plt.show()
# here people tend to buy cars with drat of range(3.0 - 3.2)
```

```python
plt.hist(cars_data['wt'])
plt.xlabel('wt')
plt.ylabel('Frequency')
plt.title('Histogram of wt')
plt.show()
# here people tend to buy cars with wt of range(3.1 - 3.8)
```

```python
plt.hist(cars_data['qsec'])
plt.xlabel('qsec')
plt.ylabel('Frequency')
plt.title('Histogram of qsec')
plt.show()
# here people tend to buy cars with qsec of range(17 - 18)
```

```python
plt.bar(cars_data['cyl'],cars_data['mpg'])
plt.xlabel('cyl')
plt.ylabel('mpg')
plt.title('Bar plot of cyl vs mpg')
plt.show()
# here people tend to buy cars with cycl 4 with mpg o
```

```python
sns.barplot(data=cars_data,x='cyl',y='mpg')
plt.title("Bar plot of cyl vs mpg")
plt.show()
#
```

```python
sns.pairplot(cars_data)
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].mean()).round().astype(int)
titanic
```

```python
titanic['deck'] = titanic['deck'].fillna("B")
titanic
```

```python
plt.hist(titanic['age'])
plt.xlabel('age')
plt.ylabel('Frequency')
plt.title('Histogram of age')
plt.show()
# People from age group of 25-32 are the ones that came out alive the most
```

```python
plt.hist(titanic['fare'])
plt.xlabel('fare')
plt.ylabel('Frequency')
plt.title('Histogram of fare')
plt.show()
#
```

```python
titanic['class'].value_counts()
```

```python
titanic['alone'] = titanic['alone'].astype(int)
titanic
```

```python
titanic['alive'] = titanic['alive'].map({'yes': 1, 'no':0})
titanic
```

```python
titanic['adult_male'] = titanic['adult_male'].astype(int)
titanic
#
```

```python
titanic['embarked'] = titanic['embarked'].fillna('S')
titanic.isna().sum()
```

```python
titanic['embark_town'].value_counts()
```

```python
titanic['embark_town'] = titanic['embark_town'].fillna('Southampton')
titanic.isna().sum()
#
```

```python
titanic
```

```python
sns.barplot(titanic,x=titanic['sex'],y=titanic['age'])
plt.xlabel('sex')
plt.ylabel('age')
plt.title('Bar plot of sex vs age')
plt.show()
```

```python
sns.barplot(titanic,x=titanic['survived'],y=titanic['age'])
plt.xlabel('survived')
plt.ylabel('age')
plt.title('Bar plot of survived vs age')
plt.show()
```
