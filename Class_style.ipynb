```python
class SecureUser:
    def __init__(self, username, pin):
        self.username = username
        self.__pin = pin
        self._balance = 0

    def set_pin(self, new_pin):
        if (new_pin == self.__pin):
            return False
        self.__pin = new_pin
        return True

    def get_balance(self):
        return self._balance
# __init__ is the constructor in a class it essential for every class
# self is required for assigning values passed by the user
# "__" of "__pin" is used so that it can't be used outside the class,it makes data private
# This mechanism is called encapsulation
```

```python
john = SecureUser('John', 1234)
john._balance = 1000
john.get_balance()
```
