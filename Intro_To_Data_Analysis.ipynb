### Finding Null values in each feature

```python
hotel_data.isnull().sum()
# .isnull() - used to check or find null values in the data set
# .sum() - used find the total number of teh null values in each feature
# .isnull().sum() - This method is called chain of functions or chain of commands
```

```python
hotel_data.describe()
# .describe() - used to generate descriptive statistics
# describe(include = "all") - include generally initialized to "none",
# if gvien value "all" it takes all features. Otherwise it only takes "int64" data types usually
```

```python
hotel_data.describe(include="all")
```

```python
hotel_data.dtypes
# .dtypes is an attribute used to find data types of features
```

```python
hotel_data.nunique()
# nunique() counts how many unique values are present in a column.
```

```python
# unique() returns all the distinct values from a column.
items = hotel_data["item_name"].unique()
for item in items:
    print(item)
len(items)
```

```python
# This code groups the data by each food item (item_name) and
# then adds up the total quantity sold for every item.
# It shows how many units of each item were ordered in total.
hotel_data.groupby("item_name")["quantity"].sum()
```

```python
hotel_data.groupby("item_name")["quantity"].sum().sort_values(ascending=False)
# sort_values arrages values in ascending order
# attribute "ascending=False" reverses order into desending order
```

```python
hotel_data.groupby("item_name")["quantity"].sum().sort_values(ascending=False).head()
# head() gives first 5 rows of the table!
```

```python
hotel_data.groupby("item_name")["quantity"].sum().sort_values().head(10)
```

```python
# .str was used to convert the data type of the selected feature 'item_price' from object to string
# initially to remove the "$" symbol infront of it!
# .replace() was used to replace "$". After that the data type will converted back to object!
# .strip() was used to remove the white spaces in the data
hotel_data['item_price'] = (
    hotel_data['item_price']
    .str.replace("$", "")
    .str.strip()
    .astype(float)
)
```

```python
hotel_data
```

```python
hotel_data.dtypes
```

```python
hotel_data.groupby("item_name")["item_price"].sum().sort_values(ascending=False).head()
```

```python
hotel_data.groupby("item_name")["item_price"].sum().sum()
```

```python
hotel_data.groupby("item_name")[["quantity","item_price"]].sum().sort_values(ascending=False,by="quantity").head()
```
